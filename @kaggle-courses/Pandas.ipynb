## DataFrame
A DataFrame is a table. It contains an array of individual entries, each of which has a certain value. Each entry corresponds to a row (or record) and a column.

````python
pd.DataFrame({
    'Y': [50, 21], 
    'N': [131, 2]
})
````

````python
pd.DataFrame({
    'Bob': ['I liked it', 'It was awful'],
    'Sue': ['Pretty good', 'Bland...']},
    index=['Product A', 'Product B']
)
````

## Series
A Series, by contrast, is a sequence of data values. If a DataFrame is a table, a Series is a list. It is, in essence, a single column of a DataFrame

````python
pd.Series([1, 2, 3, 4, 5])
````

````python
pd.Series([30, 35, 92], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
````

### Index-based selection
selects data based on its numerical position in the data.

````python
reviews.iloc[2]
````

### Label-based selection
With this type of selection method, it's the data index value, not its position, which matters.

````python
reviews.loc[0, ['country', 'points', 'title']]
````

### ```loc``` vs ```iloc```
```iloc``` uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So ```0:10``` will select entries ```0,...,9```. ```loc```, meanwhile, indexes inclusively. So ```0:10``` will select entries ```0,...,10```.

# Maps
A **map** is a term, borrowed from mathematics, for a function that takes one set of values and "maps" them to another set of values.

````python
review_points_mean = reviews.points.mean()
reviews['points'].map(lambda pts: pts-review_points_mean)
````

```apply()```: The equivalent method to transform a whole DataFrame by calling a custom method on each row

````python
def remean_points(row):
    row['points_remeaned'] = row['points'] - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
````

# Grouping and Sorting

## Groupwise analysis
```groupby()``` created a group of reviews which allotted the same point values to the given wines. Then, for each of these groups, we grabbed the ```points()``` column and counted how many times it appeared. ```value_counts()``` is just a shortcut to this ```groupby()```operation

````python
reviews.groupby('points').points.count()
````

````python
reviews.groupby('points').price.min()
````

Each group generated with ```groupby()``` can be thought as a slice of DataFrame containing only data with values that match. This DataFrame is accessible directly using the ```apply()``` method, and this then can be further manipulated. Below is an example where only the name of the first wine reveiwed from each winery is selected:

````python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
````

And here is an example where the best wine by country and province is selected (manipulating multiple columns)

````python
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
````

Another ```groupby()``` method worth mentioning is ```agg()```, which lets you run a bunch of different functions on your DataFrame simultaneously. For example, below is a simple statistical summary of the dataset:

````python
reviews.groupby(['country']).price.agg([len, min, max])
````

## Multi-indexes

````python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
````

````python
mi = countries_reviewed.index
type(mi)
````

### ```reset_index()```
In general the multi-index method which will be used most often is one for converting back to a regular index, the ```reset_index()``` method

````python
countries_reviewed.reset_index()
````

## Sorting
When outputting the result of a groupby, the order of the rows is dependent on the values in the index, not in the data. To get data in the wanted order, the ```sort_values()``` method is handy.

````python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
````

```sort_values()``` defaults to an ascending sort, where the lowest values go first. However, most of the time a descending sort is appropriate, where the higher numbers go first. That goes thusly:

````python
countries_reviewed.sort_values(by='len', ascending=False)
````

To sort by index values, use the companion method ```sort_index()```. This method has the same arguments and default order:

````python
countries_reviewed.sort_index()
````

````python
countries_reviewed.sort_values(by=['country', 'len'])
````

````python
countries_reviewed.head()
````

## Data Types and Missing Values

### Dtypes
The data type for a column in a DataFrame or a Series is known as the **dtype**

````python
reviews.price.dtype
````

Alternatively, the ```dtypes``` property returns the ```dtype``` of *every* column in the DataFrame

````python
reviews.dtypes
````

One peculiarity to keep in mind is that columns consisting entirely of strings do not get their own type; they are instead given the ```object``` type.

It's possible to convert a column of one type into another whereever such a conversion makes sense by using the ```astype()``` function. 

````python
print(reviews.points.dtype)
reviews.points.astype('float64')
````

## Combining
Pandas has three core methods for combining different DataFrames and/or Series: ```concat()```, ```join()```, and ```merge()``` (although most of what ```merge()``` can do can also be done more simply with ```join()```)

```concat()```, given a list of elements, will smush those elements together along an axis. This is useful when we have data in different DataFrame or Series objects but having the same fields (columns). One example: the [YouTube Videos dataset](https://www.kaggle.com/datasnaek/youtube-new?select=KRvideos.csv), which splits the data up based on country of origin

````python
ca_youtube = pd.read_csv('./data/CAvideos.csv')
us_youtube = pd.read_csv('./data/USvideos.csv')

pd.concat([ca_youtube, us_youtube]).head()
````

```join()``` lets you combine different DataFrame objects which have an index in common

````python
left = ca_youtube.set_index(['title', 'trending_date'])
right = us_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK').head()
````
