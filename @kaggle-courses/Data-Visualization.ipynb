```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Line Charts

```python
museum = pd.read_csv('./data/museum_visitors.csv', index_col='Date', parse_dates=True)
museum.head()
```

```python
# How many visitors did the Chinese American Museum 
# receive in July 2018?
print(f"Chinese American Museum, Jul 2018: {museum.loc['2018-07-01', 'Chinese American Museum']}")

# In October 2018, how many more visitors did Avila 
# Adobe receive than the Firehouse Museum?
print(f"Avila Adobe - Firehouse : {museum.loc['2018-10-01', 'Avila Adobe']}")
```

```python
# Line chart showing the number of visitors to each museum over time
plt.figure(figsize=(12,6))
plt.title('Monthly visitors')
sns.lineplot(data=museum)
```

```python
# Line plot showing the number of visitors to Avila Adobe over time
plt.figure(figsize=(12,6))
plt.title('# of visitors to Avila Adobe over time')
sns.lineplot(data=museum.loc[:, 'Avila Adobe'])
```

# Bar Charts and Heatmaps

```python
ign = pd.read_csv('./data/ign_scores.csv', index_col='Platform')
ign
```

```python
# What is the highest average score received by PC games,
# for any platform?
high_score = ign.loc['PC'].max()
print(f"platform w highest avg PC game score: {high_score}")

# On the Playstation Vita platform, which genre has the 
# lowest average score?
worst_genre = ign.loc['PlayStation Vita'].idxmin()
print(f"lowest rated genre on Playstation Vita: {worst_genre}")
```

```python
# Bar chart showing average score for racing games by platform
plt.figure(figsize=(8,6))
plt.title('Average score for racing games per platform')
sns.barplot(data=ign, y=ign.index, x='Racing')
```

```python
# Heatmap showing average game score by platform and genre
plt.figure(figsize=(10,8))
sns.heatmap(data=ign, annot=True)
```

# Scatter Plots

```python
candy = pd.read_csv('./data/candy.csv', index_col='id')
candy.head()
```

```python
# Which candy was more popular with survey respondents:
# '3 Musketeers' or 'Almond Joy'?
almond_joy_score = float(candy.loc[candy['competitorname']=='Almond Joy']['winpercent'])
musketeers_score = float(candy.loc[candy['competitorname']=='3 Musketeers']['winpercent'])
print("Survey Score")
print(f"Almond Joy: {almond_joy_score} \n3Musketeers: {musketeers_score}")
```

```python
# Which candy has higher sugar content: 'Air Heads'
# or 'Baby Ruth'?
air_heads_sugar = round(float(candy.loc[candy['competitorname']=='Air Heads']['sugarpercent'])*100)
baby_ruth_sugar = round(float(candy.loc[candy['competitorname']=='Baby Ruth']['sugarpercent'])*100)
print("Sugar Content")
print(f"Air Heads: {air_heads_sugar}% \nBaby Ruth: {baby_ruth_sugar}%")
```

## scatterplot, regplot, lmplot

```python
# Scatter plot showing the relationship between 'sugarpercent' and 'winpercent'
sns.scatterplot(x=candy.sugarpercent, y=candy.winpercent)
```

```python
# Scatter plot w/ regression line showing the relationship between 'sugarpercent' and 'winpercent'
sns.regplot(x=candy.sugarpercent, y=candy.winpercent)
```

```python
# Scatter plot showing the relationship between 'pricepercent', 'winpercent', and 'chocolate'
sns.scatterplot(x=candy.pricepercent, y=candy.winpercent, hue=candy.chocolate)
```

```python
sns.lmplot(data=candy, x= 'pricepercent', y= 'winpercent', hue= 'chocolate')
```

## swarmplot

```python
# Scatter plot showing the relationship between 'chocolate' and 'winpercent'
sns.swarmplot(x=candy.chocolate, y=candy.winpercent)
```
