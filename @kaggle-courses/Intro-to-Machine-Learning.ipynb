````python
import pandas as pd

melbourne_data = pd.read_csv('./data/melb_data.csv')
melbourne_data.head()
````

````python
melbourne_data.describe()
````

````python
melbourne_data.dropna(axis=0, inplace=True)
melbourne_data.describe()
````

## Selecting the Prediction Target

````python
y = melbourne_data.Price
y.head()
````

## Choosing 'Features'

````python
features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = melbourne_data[features]
X.describe()
````

````python
X.head()
````

## Building a Model
The steps to building and using a model are:

* **Define**: What type of model will it be? A decision tree? Some other type of model? Some other parameters of the model type are specified too.
* **Fit**: Capture patterns from provided data. This is the heart of modeling.
* **Predict**: Just what it sounds like
* **Evaluate**: Determine how accurate the model's predictions are.

````python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=7)
model.fit(X,y)
````

````python
print("Making predictions for the first 10 houses...")
print(X.head())

print("\nPrediction Values:")
print(model.predict(X.head()))
````

````python
sample_predictions = list(model.predict(X.head()))
actual_prices = list(melbourne_data['Price'][:5])

for i in range(5):
    print(f"House #{i+1}:")
    print(f"Predicted Price: {sample_predictions[i]}")
    print(f"Actual Price: {actual_prices[i]}\n")
````

## Model Validation
There are many metrics for summarizing model quality, one of which is called **Mean Absolute Error** (also called **MAE**). The MAE metric takes the absolute value of each error (```error=actual-predicted```) and the average of those absolute errors is calculated. In plain English, it can be said as 
> On average, the predictions are off by about X

````python
from sklearn.metrics import mean_absolute_error

predictions = model.predict(X)
mean_absolute_error(y, predictions)
````

````python
# spliting data into training and validation data 
# (for both features and target)
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=7)

model = DecisionTreeRegressor()
model.fit(train_X, train_y)
````

````python
print(train_X.shape, val_X.shape, train_y.shape, val_y.shape)
````

````python
val_predictions = model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
````

## Optimizing Models

* **overfitting**: where a model matches the training data almost perfectly, but does poorly in validation and other new data. Overfitting captures spurious patterns that won't recur in the future, leading to less accurate predictions
* **underfitting**: When a model fails to capture important or relevant distinctions and patterns in the data, so it performs poorly even in training data


````python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=7)
    model.fit(train_X, train_y)
    predictions = model.predict(val_X)
    mae = mean_absolute_error(val_y, predictions)
    return(mae)
````

````python
# Comparing different tree sizes
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]
scores = {leaf_size: get_mae(leaf_size, train_X, val_X, train_y, val_y) for leaf_size in candidate_max_leaf_nodes}
print(scores)

best_tree_size = min(scores, key=scores.get)
````

````python
# Fit model using all data
final_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=7)
final_model.fit(X, y)
````

# Random Forests
A deep decision tree with lots of leaves will overfit because each prediction is coming from historical data from only the few data at its leaf. But a shallow tree with few leaves will perform poorly because it fails to capture as many distinctions in the raw data.

Even today's most sophisticated modeling techniques face this tension between underfitting and overfitting. But, many models have clever ideas that can lead to better performance. 

The **random forest** uses many trees, and it makes a prediction by averaging the predictions of each component tree. It generally has much better predictive accuracy than a single decision tree and it works well with default parameters.

````python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state=7)
forest_model.fit(train_X, train_y)
forest_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, forest_preds))
````
